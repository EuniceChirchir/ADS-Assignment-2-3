##### Breakdown of distance from home by job role and attrition.

```python
df.groupby(['JobRole']).count()
```

```python
#selecting the three columns
selection=df.loc[:1469,['Attrition','JobRole','DistanceFromHome']]
print(selection)
```

```python
df.groupby(['JobRole','Attrition'])['DistanceFromHome'].mean().unstack()
```

```python
#Data visualization of distance from home by job role and attrition
sns.catplot(x='JobRole',y='DistanceFromHome',hue='Attrition',kind='bar',data=df)
plt.xticks(rotation=90)
```

```python
sns.catplot(x='Attrition',hue='Attrition',kind='count',data=df)
```
